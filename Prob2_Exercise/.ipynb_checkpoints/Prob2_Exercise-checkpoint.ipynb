# Exercise P2.2:  Bayesian models in everday cognition

In a study that will only be disclosed in the solutions of this exercises, people were asked to make predictions about the duration or extent of everyday phenoma. Here, we look at the following two predictions that people were asked to make:

1. Suppose there is a movie that has made 50 million dollars at the box office so far. You don't know for how long the movie has been in the cinemas. What would be your guess for the total amount of money that the movie will make?

2. Predict a person's age at which they will die given their current age.
